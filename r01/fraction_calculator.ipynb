```python
'''Application that performs operations on fractions.'''

from fractions import Fraction

def print_menu():
    print("What operation do you want to perform?")
    print("1. Addition of fractions")
    print("2. Subtraction of fractions")
    print("3. Multiplication of fractions")
    print("4. Division of fractions")
    print("5. Powering up fractions")
    print("6. Extracting  the root of a fraction")
    print("7. End")
    print(flush=True)

def get_fraction():
    return Fraction(input("Specify the fraction: "))

def get_number():
    return int(input("Specify the number: "))

def add_fractions(a, b):
    return a + b

def subtract_fraction(a, b):
    return a - b

def multiply_fractions(a, b):
    return a * b

def divide_fractions(a, b):
    return a / b

def power_up_fractions(a, b):
    return a ** b

def root_of_fraction(a, b):
    return a ** (1 / b)


if __name__ == '__main__':
    while True:
        print_menu()
        try:
            choice = int(input("Enter the number corresponding to your choice: "))

            if choice == 1:  # addition
                a = get_fraction()
                b = get_fraction()
                score = add_fractions(a, b)
                print(f"{a} + {b} = {score}")
                print(flush=True)

            elif choice == 2:  # subtraction
                a = get_fraction()
                b = get_fraction()
                score = subtract_fraction(a, b)
                print(f"{a} - {b} = {score}")
                print(flush=True)

            elif choice == 3:  # multiplication
                a = get_fraction()
                b = get_fraction()
                score = multiply_fractions(a, b)
                print(f" {a} * {b} = {score}")
                print(flush=True)

            elif choice == 4:  # division
                a = get_fraction()
                b = get_fraction()
                score = divide_fractions(a, b)
                print(f"{a} / {b} = {score}")
                print(flush=True)

            elif choice == 5:  # power
                a = get_fraction()
                b = get_number()
                score = power_up_fractions(a, b)
                print(f"{a} ** {b} = {score}")
                print(flush=True)

            elif choice == 6:  # root
                a = get_fraction()
                b = get_number()
                score = root_of_fraction(a, b)
                print(f"{a} ** 1 / {b} = {score}")
                print(flush=True)

            elif choice == 7:  # end
                break

            else:  # invalid number 
                print("Invalid selection. Please try again.")

        except ValueError:
            print('An invalid value was specified. Please try again.')
            print(flush=True)

        except ZeroDivisionError:
            print("Division by zero or invalid root operation. Please try again.")
            print(flush=True)
```
