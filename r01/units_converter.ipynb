```python
''' A program that converts units of the metric system to imperial and vice versa.

    Supported conversion methods:
    Kilometers to miles / Miles to kilometers
    Kilograms to pounds / Pounds to kilograms
    Celcius degrees to Fahrenheit / Fahrenheit degrees to Celcius '''

def category_choice_menu():
    print("Select a conversion category: ")
    print("1. Measure")
    print("2. Mass")
    print("3. Temperature")
    print("4. End program")
    print(flush=True)

def category_choice():
    return int(input("Select the category: "))

def measure_converter_menu():
    print("Select type of conversion: ")
    print("1. Kilometers to miles")
    print("2. Miles to kilometers")
    print("3. Return")
    print(flush=True)
    
    return int(input("Select conversion type: "))

def mass_converter_menu():
    print("Select type of conversion: ")
    print("1. Kilograms to pounds")
    print("2. Pounds to kilograms")
    print("3. Return")
    print(flush=True)

    return int(input("Select conversion type: "))

def temperature_conversion_menu():
    print("Select type of conversion: ")
    print("1. Celcius degrees to Fahrenheit")
    print("2. Fahrenheit degrees to Celcius")
    print("3. Return")
    print(flush=True)

    return int(input("Select conversion type: "))

def km_miles_conversion(km):
    return round(km / 1.609, 3)

def miles_km_conversion(miles):
    return round(miles * 1.609, 3)

def kg_pounds_conversion(kg):
    return round(kg * 2.2046, 3)

def pounds_kg_conversion(pounds):
    return round(pounds / 2.2046, 3)

def celsius_fahrenheit_conversion(c):
    return round((c * 9/5) + 32, 3)

def fahrenheit_celsius_conversion(f):
    return round((f - 32) * 5/9, 3)

if __name__ == '__main__':
    while True:
        try:
            category_choice_menu()
            cat_choice = category_choice()

            if cat_choice == 1:  # measure unit conversion
                conv_choice = measure_converter_menu()

                if conv_choice == 1:  # km -> miles
                    distance = float(input("Enter the distance in kilometers: "))
                    miles = km_miles_conversion(distance)
                    print(f'The specified distance: {distance} km, is equal to {miles} miles.')

                elif conv_choice == 2:  # miles -> km
                    distance = float(input("Enter the distance in miles: "))
                    km = miles_km_conversion(distance)
                    print(f'The specified distance: {distance} miles, is equal to {km} kilometers.')
                
                elif conv_choice == 3:  # return to category choice
                    continue

                else:  # invalid number
                    print("Invalid selection. Please try again.")

            elif cat_choice == 2:  # mass unit conversion
                conv_choice = mass_converter_menu()

                if conv_choice == 1:  # kg -> pounds
                    weight = float(input("Enter weight in kilograms: "))
                    pounds = kg_pounds_conversion(weight)
                    print(f'The specified weight: {weight} kg, is equal to {pounds} pounds.')
                    
                elif conv_choice == 2:  # pounds -> kg
                    weight = float(input("Enter weight in pounds: "))
                    kg = pounds_kg_conversion(weight)
                    print(f'The specified weight: {weight} pounds, is equal to {kg} kg.')
                    
                elif conv_choice == 3:  # return to category choice
                    continue

                else:  # invalid number
                    print("Invalid selection. Please try again.")

            elif cat_choice == 3:  # temperature units conversion
                conv_choice = temperature_conversion_menu()

                if conv_choice == 1:  # Celsius -> Fahrenheit
                    temp = float(input("Enter temperature in Celsius degrees: "))
                    f = celsius_fahrenheit_conversion(temp)
                    print(f'The specified temperature: {temp} degrees Celsius, is equal to {f} degrees Fahrenheit.')

                elif conv_choice == 2:  # Fahrenheit -> Celsius
                    temp = float(input("Enter temperature in Fahrenheit degrees: "))
                    c = fahrenheit_celsius_conversion(temp)
                    print(f'The specified temperature: {temp} degrees Fahrenheit, is equal to {c} degrees Celsius.')
                    # tutaj funkcja poboru Fahrenheit√≥w i przeliczenie na Celcjusze
                elif conv_choice == 3:  # return to category choice
                    continue

                else:  # invalid number
                    print("Invalid selection. Please try again.")
                    
            elif cat_choice == 4:  # end the program
                break
        except ValueError:
            print('An invalid value was specified. Please try again.')    
```
